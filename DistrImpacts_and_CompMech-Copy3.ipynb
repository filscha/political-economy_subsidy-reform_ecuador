## Adjust monetary expenditures to match official numbers and for durable goods

```python
lifetime = 10 # INPUT
# Create new DataFrame with relevant variables
expend = hh_expend[["hh_ID", "weight", "expend_item_no", "expend_mon"]].fillna(0)
# Combine dataset with respective IOT commodities and expenditure categories
expend = expend.join(expend_IOT.set_index("expend_item_no"), on = "expend_item_no")

# Calculate weighted expenditures
expend["expend_mon_w"]=expend["expend_mon"].mul(expend["weight"], axis=0)

# Set expenditures for agricultural activities and property purchases equal to 0 (as in official numbers)
expend[["expend_mon_noAP","expend_mon_w_noAP"]] = expend[["expend_mon","expend_mon_w"]] #create new columns
agro_property = (hh_expend.type == 7) | (hh_expend.expend_item_no == 1801001) | (hh_expend.expend_item_no == 1801002) | (hh_expend.expend_item_no == 1801003) | (hh_expend.expend_item_no == 1801004) #filter
expend.loc[agro_property, ["expend_mon_noAP","expend_mon_w_noAP"]] = 0 #Set corresponding values equal to 0
# Check: Compare difference w/ official data
check_agro_property = round(expend["expend_mon_w"].sum() - expend["expend_mon_w_noAP"].sum(),4) == round((hh_expend[agro_property]["weight"]*hh_expend[agro_property]["expend_mon"]).sum(),4)

# Discount durable goods over lifetime
expend[["expend_mon_noAP_dd","expend_mon_w_noAP_dd"]] = expend[["expend_mon_noAP","expend_mon_w_noAP"]] #create new columns
durables_no = categ_no_label.loc[categ_no_label["Category_label"] == "Durable goods", "Category_no"].iloc[0] #get Category_no for durables
expend.loc[expend.Category_no == durables_no, ["expend_mon_noAP_dd","expend_mon_w_noAP_dd"]] = expend.loc[expend.Category_no == durables_no, ["expend_mon_noAP_dd","expend_mon_w_noAP_dd"]]/lifetime #discount expenditures
#Check: 
#expend.loc[expend.Category_no == durables].head(5) #show durable goods
durables_mon = expend.loc[expend.Category_no == durables_no][["expend_mon_w_noAP"]].sum().sum()
check_durables = round(expend["expend_mon_w_noAP"].sum()- expend["expend_mon_w_noAP_dd"].sum(),4) == round(durables_mon-durables_mon/10,4)

# "expend_mon_w_noAP_dd" is used to model additional expenditures and instead of "gas_gru_cor" to get proxy for income

print("The variable expend_mon_w_noAP represents the official reported number. \n",
      "The variable expend_mon_w_noAP_dd is adjusted for durable goods (discounted) and used to model expenditure increases and to replace the aggregated variable gas_gru_cor for total expenditures as income proxy.\n",
     "The sanity check yields:", check_durables & check_agro_property)
```

## Input-output price model
To assess impact on sector prices due to energy price hike

### Prepare IOT

```python
Z = IOT.iloc[:245, :245] # Square matrix of intermediate flows
V = IOT.iloc[246, 0:245].values + IOT.iloc[250, 0:245].values # Value added (wages, interest, profit) plus taxes
M = IOT.iloc[248, 0:245].values # imports
X = IOT.iloc[251, 0:245].values # Total production output X
```


### Aggregate Fuel oil, and Other fuels to Diesel sector

```python
# INPUT: IOT_no of Fuel oil and Other fuels
Dsl_agg1_no = 132; Dsl_agg2_no = 133;
Dsl_agg1_name = dict_IOT_no_label.get(Dsl_agg1_no); Dsl_agg2_name =  dict_IOT_no_label.get(Dsl_agg2_no) #Names of to be aggregated

# Aggregate Z
Z = Z.rename(index={Dsl_agg1_name:"Diesel" , Dsl_agg2_name:"Diesel"}, columns={Dsl_agg1_name:"Diesel" , Dsl_agg2_name:"Diesel"}) #Rename to be aggregated to aggregate
Z = Z.groupby(Z.index).sum() #Groupby index, aggregating the Diesel rows
Z = Z.T.groupby(Z.T.index).sum().T #Transpose and groupby index, aggregating the Diesel columns
Z = Z.rename(index = dict_IOT_label_no, columns = dict_IOT_label_no).sort_index().sort_index(axis=1) #Put in original order
Z = Z.rename(index = dict_IOT_no_label, columns = dict_IOT_no_label) #Put IOT labels instead of numbers
# Aggregate V
V[[dict_IOT_label_no.get("Diesel")-1]] = V[[dict_IOT_label_no.get("Diesel")-1,Dsl_agg1_no-1,Dsl_agg2_no-1]].sum() #Aggregate to Diesel (#Index starts at 0 --> -1)
V = np.delete(V, [Dsl_agg1_no-1,Dsl_agg2_no-1]) #Delete aggregated values
# Aggregate M
M[[dict_IOT_label_no.get("Diesel")-1]] = M[[dict_IOT_label_no.get("Diesel")-1,Dsl_agg1_no-1,Dsl_agg2_no-1]].sum() #Aggregate to Diesel (#Index starts at 0 --> -1)
M = np.delete(M, [Dsl_agg1_no-1,Dsl_agg2_no-1]) #Delete aggregated values
# Aggregate X
X[[dict_IOT_label_no.get("Diesel")-1]] = X[[dict_IOT_label_no.get("Diesel")-1,Dsl_agg1_no-1,Dsl_agg2_no-1]].sum() #Aggregate to Diesel (#Index starts at 0 --> -1)
X = np.delete(X, [Dsl_agg1_no-1,Dsl_agg2_no-1]) #Delete aggregated values

# Doublecheck if the sum of intermediate and primary inputs is equal total production with a tolerance of 0.01%
check = np.allclose(Z.sum(axis = 0).values + V + M, X, rtol=0.01)

#Check
print(Dsl_agg1_name, "and", Dsl_agg2_name, "were aggregated to the Diesel sector. Check yields:", check)
```

### Leontief inverse and price change vector for scenarios

```python
# Calculate matrix of technical coefficients A by dividing intermediate inputs of IOT with total output X
A = Z/X; A_t = A.transpose()

# Calculate Leontief inverse
I = np.identity(len(A)) # Identity matrix
L_array = np.linalg.inv((I-A)) # Leontief inverse in numpy.ndarray
L = pd.DataFrame(L_array, index = A.index, columns = A.columns) # Leontief inverse in Pandas.DataFrame

# Extract rows with relevant energy types
IOT_price = L[L.index==""].copy()
for en in list(energy.keys()):
    IOT_price_en = L[L.index==en]
    IOT_price = pd.concat([IOT_price, IOT_price_en], axis=0)
IOT_price = IOT_price.T #Transpose

# Price change vector for different scenarios
IOT_price_delta = pd.DataFrame(data=None, columns=[""], index=IOT_price.index)
no_sc = price_scenarios.shape[1] #number of scenarios
for x, en in enumerate(list(energy.values())): # loop through relevant energy types
    for y in range(no_sc): # loop through scenarios (stored in columns of dataframe)
        IOT_price_delta.insert(loc=(x*no_sc+y), column=en+"_"+price_scenarios.columns[y], value=IOT_price[list(energy.keys())[x]]*price_scenarios.iloc[x,y])
del IOT_price_delta[""]
IOT_price_delta = IOT_price_delta.rename(index = dict_IOT_label_no) #Replace IOT labels with numbers
en_sc = list(IOT_price_delta.columns.values) # list with energy types and scenarios (for headers)
```

### Expand existing cash transfers to current non-beneficiaries

```python
comp_name = "Expanding BDH"
BDH_monthly = 50 #Specify monthly payment of existing cash transfer

# DESIGN
# Expand BDH to non-receiving hh up to certain income_pc (100% revenue-recycling)
#DataFrame with hh that don't receive BDH
income_pc_hh = master.groupby("hh_ID")[["weight","income_pc"]].mean().reset_index() #income_pc per hh
BDH_non_hh = income_pc_hh[~income_pc_hh["hh_ID"].isin(list((BDH_hh["hh_ID"])))].sort_values("income_pc").reset_index(drop=True) #Select only those hh that don't receive the BDH and sort by per capita income
BDH_non_hh["weight_cum"] = BDH_non_hh["weight"].cumsum() #Cumulative sum of weights to know real no of hh

# Number of hh
BDH_expand = revenues.copy(); BDH_expand["No_hh_add"] = revenues/BDH_monthly #Get max number of additional hh that receive BDH
for x in range(len(BDH_expand.index.values.tolist())): #Loop through energy scenarios and calculate maximum income_pc of hh that could be financed by revenue-recycling
    BDH_expand.loc[BDH_expand.index.values.tolist()[x],"income_pc_max"] = \
    BDH_non_hh.iloc[(BDH_non_hh["weight_cum"]-BDH_expand.iloc[x,1]).abs().argsort()[:1]]["income_pc"].tolist()[0]
BDH_expand = BDH_expand.T #Transpose

# RESULTS
# Additional income as incomeshare from BDH expansion by income quintiles
income_add_BDHexp_quint = pd.DataFrame(index=quintile_labels); income_add_BDHexp_quint.index.name = "quintile"

for x, item in enumerate(en_sc):
    BDH_NewRecip = pd.DataFrame(BDH_non_hh[BDH_non_hh["income_pc"]<=BDH_expand.iloc[2,x]]["hh_ID"]) # Get households ID's that receive the expanded BDH
    BDH_NewRecip[item +"_"+ comp_name] = BDH_monthly
    income_hh_BDHexp = income_hh_BDH.merge(BDH_NewRecip, how="outer", on="hh_ID") #Merge new BDH recipients to income dataset
    income_hh_BDHexp[item +"_"+ comp_name].fillna(0, inplace=True) #replace missing values with 0 (hh w/o BDH expand)
    income_hh_BDHexp[list(income_hh_BDHexp)[-1]+"_incomeshare"]=income_hh_BDHexp[list(income_hh_BDHexp)[-1]]/income_hh_BDHexp["income_hh"] #incomeshares
    income_hh_BDHexp[list(income_hh_BDHexp)[-1]+"_w"] = income_hh_BDHexp[list(income_hh_BDHexp)[-1]].mul(income_hh_BDHexp["weight"], axis=0) #weighted income share
    income_hh_BDHexp = pd.DataFrame(data=income_hh_BDHexp.groupby("quintile")[list(income_hh_BDHexp)[-1]].sum().div(no_hh_quintiles, axis=0), columns=[list(income_hh_BDHexp)[-1]]) #groupby income quintiles
    income_add_BDHexp_quint = pd.concat([income_add_BDHexp_quint, income_hh_BDHexp], axis=1) #collect for all scenarios

# Net additional income as percentage share of income
net_income_add_BDHexp_quint = income_add_BDHexp_quint.T.reset_index(drop=True).T - results_quint_distr[col_total_distr_all].T.reset_index(drop=True).T
net_income_add_BDHexp_quint.columns = ["net_" + s for s in list(income_add_BDHexp_quint.columns.values)]

# Remainder is 0
BDHexpNB_remainder = pd.Series(data=0, index=en_sc)

# Store data for graphs in dictionaries
income_add_quint[comp_name] = income_add_BDHexp_quint
net_income_add_quint[comp_name] = net_income_add_BDHexp_quint
remain[comp_name] = BDHexpNB_remainder
if comp_name not in comp_list: comp_list.append(comp_name) 
income_legend[comp_name]="Income from expanded BDH to non-beneficiaries"
```

### Pension for everyone above 65 years

```python
comp_name = "Minimum Pension"
age = 65

# DESIGN
# Source Dataframe
pers_income_pension_age = pers_income_pension[pers_income_pension["Age"]>= age].copy().reset_index() #Select only those persons above retirement age

# Determine minimum and maximum minimum Pension
no_pers_NoPension = pers_income_pension_age[pers_income_pension_age["Pension"]==0]["weight"].sum()
MP_simple = revenues/no_pers_NoPension
MP_min = round(MP_simple["revenues_monthly"].min(),0)
MP_max = int(MP_simple["revenues_monthly"].max())+10

# Costs of different minimum pensions
pers_income_pension_MP = pers_income_pension_age[["pers_ID", "weight", "Pension"]].copy()
pers_income_pension_MP = pers_income_pension_MP.sort_values("Pension")
MP = MP_min
while MP < MP_max:
    pers_income_pension_MP[MP] = np.where(pers_income_pension_MP["Pension"]<=MP, MP - pers_income_pension_MP["Pension"], 0)
    MP +=.5
min_col = pers_income_pension_MP.columns.get_loc(MP_min)
pers_income_pension_MP_costs = pd.DataFrame(pers_income_pension_MP.iloc[:,min_col:].mul(pers_income_pension_MP["weight"], axis=0).sum())
pers_income_pension_MP_costs = pers_income_pension_MP_costs.rename(columns = {0:"costs"})
pers_income_pension_MP = None #Clear memory

# Affordable minimum pension
MP_real = pd.DataFrame(columns = [comp_name, "costs"])
for i in range(revenues.shape[0]): #Loop through each row in revenus dataframe and check which MP can be financed
    MP_real.loc[i] = list(pers_income_pension_MP_costs.iloc[(pers_income_pension_MP_costs['costs']-revenues.iloc[i,0]).abs().argsort()[:1]].reset_index().iloc[0,:])
MP_real = pd.concat([MP_real, revenues.reset_index()], axis=1)
MP_real["revenues_remain"] = MP_real["revenues_monthly"]-MP_real["costs"]

# Assign MP to retirees
for x in range(len(en_sc)): #Product of energy types and scenarios (en_sc list)
    MP = MP_real.loc[x,comp_name]
    for i in range(pers_income_pension_age.shape[0]): #Loop through retirees
        P = pers_income_pension_age.loc[i,"Pension"] #Income from pension payments
        if P<MP: #If retirees have a current pension below minimum pension, they get the difference
            pers_income_pension_age.loc[i,en_sc[x]+"_"+comp_name+str(MP)] = MP-P
        else: #otherwise they don't get minimum pension
            pers_income_pension_age.loc[i,en_sc[x]+"_"+comp_name+str(MP)] = 0

# Groupby households and get variables
min_col = pers_income_pension_age.columns.get_loc(MP_real.loc[0,"index"]+"_"+comp_name+str(MP_real.loc[0,comp_name]))
income_hh_MP = pers_income_pension_age.groupby("hh_ID")[list(pers_income_pension_age.iloc[:,min_col:])].sum().reset_index() #Collapse by household
income_hh_MP = income_hh_MP.merge(income_hh_BDH[["hh_ID","quintile", "income_hh", "weight"]], on="hh_ID", how="left") #Add income quintiles and income 

# RESULTS
# Additional income as incomeshare from minimum pension by income quintiles
income_add_MP_quint = pd.DataFrame(index=quintile_labels); income_add_MP_quint.index.name = "quintile"
for x, item in enumerate(en_sc):
    col_name = list(pers_income_pension_age.iloc[:,min_col:])[x]
    income_hh_MP[col_name+"_incomeshare"]=income_hh_MP[col_name]/income_hh_MP["income_hh"] #incomeshares
    income_hh_MP[list(income_hh_MP)[-1]+"_w"] = income_hh_MP[list(income_hh_MP)[-1]].mul(income_hh_MP["weight"], axis=0) #weighted income share
    income_hh_MP_quint = pd.DataFrame(data=income_hh_MP.groupby("quintile")[list(income_hh_MP)[-1]].sum().div(no_hh_quintiles, axis=0), columns=[list(income_hh_MP)[-1]]) #groupby income quintiles
    income_add_MP_quint = pd.concat([income_add_MP_quint, income_hh_MP_quint], axis=1) #collect for all scenarios

# Net additional income as percentage share of income
net_income_add_MP_quint = income_add_MP_quint.T.reset_index(drop=True).T - results_quint_distr[col_total_distr_all].T.reset_index(drop=True).T
net_income_add_MP_quint.columns = ["net_" + s for s in list(income_add_MP_quint.columns.values)]

# Remainder is 0
MP_remainder = pd.Series(data=0, index=en_sc)

# Store data for graphs in dictionaries
income_add_quint[comp_name] = income_add_MP_quint
net_income_add_quint[comp_name] = net_income_add_MP_quint
remain[comp_name] = MP_remainder
if comp_name not in comp_list: comp_list.append(comp_name) 
income_legend[comp_name]="Income from minimum pension"
```

```python
comp_name = "Public transport voucher"

# DESIGN: 1. idea Compensate as much as Q2 hh who use PT have as indirect additional transportation expenditures for diesel scenario
PT_add= total["Diesel"].copy() # Total additional expenditures due to diesel SR
PT_add = PT_add[(PT_add["Category_no"]==13) | (PT_add["Category_no"]==14)] #Additional direct transport expenditures by bus
# Unweighted indirect additional transportation expenditures
PT_add[["Diesel_min_total_expend_add","Diesel_med_total_expend_add","Diesel_max_total_expend_add"]]= \
PT_add[["Diesel_min_total_expend_add_w","Diesel_med_total_expend_add_w","Diesel_max_total_expend_add_w"]].div(PT_add["weight"], axis=0)
# Weighted average indirect additional transportation expenditures of Q2
voucher_PT_add =list(np.average(PT_add[PT_add["quintile"]=="Q2"][["Diesel_min_total_expend_add","Diesel_med_total_expend_add","Diesel_max_total_expend_add"]], weights = PT_add[PT_add["quintile"]=="Q2"]["weight"], axis=0))
#--> Additional expenditures are very low --> Other basis

# 2. idea
# Compensate as much as average Q2 household has as additional expenditures
voucher_transport = round(results_quint_expend_hh.iloc[1,:12],2).to_frame(name="voucher_transport_maxUSD/hh"); voucher_transport.index=en_sc

# Expenditures for public transportation by bus
PT_bus = (master["Category_no"]==13) | (master["Category_no"]==14)
expend_PT = master[PT_bus].groupby("hh_ID")[["expend_mon_noAP_dd", "expend_mon_w_noAP_dd"]].sum().reset_index()
expend_PT = expend_PT.rename(columns={"expend_mon_noAP_dd":"expend_bus","expend_mon_w_noAP_dd":"expend_bus_w" })
expend_PT = expend_PT.merge(income_hh_BDH[["hh_ID","quintile", "income_hh","weight"]], on="hh_ID", how="left") #Add income quintiles and income 

# Distribute voucher to hh from Q1 and Q2 but only up to already existing expenditures from public transportation by bus (otherwise no need assumed)
expend_PT["quintile"]=expend_PT["quintile"].map({"Q1":1,"Q2":2,"Q3":3,"Q4":4,"Q5":5})
for x in range(len(en_sc)): #Product of energy types and scenarios (en_sc list)
    expend_PT[en_sc[x]+"_"+comp_name] =  np.where(expend_PT["quintile"]<=2, np.where(expend_PT["expend_bus"]<=voucher_transport.iloc[x,0], expend_PT["expend_bus"], voucher_transport.iloc[x,0]), 0)

# RESULT
# Additional income as incomeshare from minimum pension by income quintiles
income_hh_PT = expend_PT.copy(); expend_PT= None; income_hh_PT["quintile"]=income_hh_PT["quintile"].map({1:"Q1",2:"Q2",3:"Q3",4:"Q4",5:"Q5"})
income_add_PT_quint = pd.DataFrame(index=quintile_labels); income_add_PT_quint.index.name = "quintile"
min_col = income_hh_PT.columns.get_loc(en_sc[0]+"_"+comp_name)

for x, item in enumerate(en_sc):
    col_name = list(income_hh_PT.iloc[:,min_col:])[x]
    income_hh_PT[col_name+"_incomeshare"]=income_hh_PT[col_name]/income_hh_PT["income_hh"] #incomeshares
    income_hh_PT[list(income_hh_PT)[-1]+"_w"] = income_hh_PT[list(income_hh_PT)[-1]].mul(income_hh_PT["weight"], axis=0) #weighted income share
    income_hh_PT_quint = pd.DataFrame(data=income_hh_PT.groupby("quintile")[list(income_hh_PT)[-1]].sum().div(no_hh_quintiles, axis=0), columns=[list(income_hh_PT)[-1]]) #groupby income quintiles
    income_add_PT_quint = pd.concat([income_add_PT_quint, income_hh_PT_quint], axis=1) #collect for all scenarios

# Net additional income as percentage share of income
net_income_add_PT_quint = income_add_PT_quint.T.reset_index(drop=True).T - results_quint_distr[col_total_distr_all].T.reset_index(drop=True).T
net_income_add_PT_quint.columns = ["net_" + s for s in list(income_add_PT_quint.columns.values)]

# Remainder for public budget
PT_costs = (income_hh_PT[list(income_hh_PT.iloc[:,min_col:min_col+len(en_sc)])].mul(income_hh_PT["weight"],axis=0)).sum()
PT_remainder = (revenues.reset_index().iloc[:,1]-PT_costs.reset_index().iloc[:,1])*12/(10**6); PT_remainder.index = en_sc  #what's left for public budget in mn USD per year

# Store data for graphs in dictionaries
income_add_quint[comp_name] = income_add_PT_quint
net_income_add_quint[comp_name] = net_income_add_PT_quint
remain[comp_name] = PT_remainder
if comp_name not in comp_list: comp_list.append(comp_name) 
income_legend[comp_name]="Income from public transportation vouchers"
```
